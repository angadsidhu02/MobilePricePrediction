```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/train.csv')
df = pd.DataFrame(data)
df.head()
```

```python
for column in df.columns:
    value_counts = df[column].value_counts()
    print(f"Value counts for {column}:\n{value_counts}\n")
```

```python
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler

columns_to_scale = ['battery_power' , 'clock_speed' , 'fc' , 'int_memory' , 'm_dep' , 'mobile_wt' , 'n_cores' , 'pc' , 'px_height' , 'px_width' , 'ram', 'sc_h' , 'sc_w' , 'talk_time']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), columns_to_scale)
    ],
    remainder='passthrough'  # Pass through columns not scaled
)

# Fit and transform the data using the preprocessor
scaled_data = preprocessor.fit_transform(df)

# Create a new DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

scaled_df.drop(['blue', 'sc_h', 'sc_w', 'talk_time'], axis=1, inplace=True)

scaled_df.head()
```

```python
X = scaled_df.drop('price_range', axis= 1)
Y = scaled_df['price_range']
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state= 42)

lr = LogisticRegression()

lr.fit(X_train, Y_train)
lr.score(X_test, Y_test)
```

```python
y_pred = lr.predict(X_test)

y_pred = pd.DataFrame(y_pred)
y_pred
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y_test, y_pred)
print(mse)
```
